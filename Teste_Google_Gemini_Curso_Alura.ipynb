```python
df = pd.DataFrame(docs)
df = df.explode('Conteudo').reset_index(drop=True)
```

```python
df.head()
```

```python
def embed_fn(title,text,model='models/embedding-001'):
  return genai.embed_content(title=title,
                             content=text,
                             model=model,
                             task_type='RETRIEVAL_DOCUMENT')['embedding']
```

```python
df['Embeddings'] = df.apply(lambda row: embed_fn(row['Titulo'],row['Conteudo']),axis=1)
```

```python
df.head()
```

```python
def gerar_e_buscar_consulta(consulta,df,model):
  embedding_consulta = genai.embed_content(content=consulta,
                                           model=model,
                                           task_type='RETRIEVAL_QUERY')['embedding']
  produtos_escalares = np.dot(np.stack(df['Embeddings']),embedding_consulta)
  indice = np.argmax(produtos_escalares)
  return df.iloc[indice]["Titulo"]
```

```python
consulta = "recebi um email dizendo que minha conta foi comprometida, isso é golpe?"

tipo_golpe = gerar_e_buscar_consulta(consulta,df,model)
tipo_golpe
```

```python
prompt = f"o usuario prossivelmente esta recebendo um golpe do tipo {tipo_golpe}, explique para ele resumidamente, de forma simples, o que é esse golpe e como ele pode agir pra evita-lo"
model_2 = genai.GenerativeModel('gemini-1.0-pro')
response = model_2.generate_content(prompt)
print("Este é um padrão de golpe do tipo", tipo_golpe,"\n\n Entenda melhor sobre ele abaixo:\n\n", response.text)
```
